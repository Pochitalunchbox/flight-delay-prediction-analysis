```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# For probability density function
def f(x):
    return 0.5 * np.exp(-np.abs(x))
```

```python
# Creating metropolis-Hastings algorithm
def metropolis_hastings(N, s):
    samples = np.zeros(N)
    x = 0  
    for i in range(1, N):
        x_star = np.random.normal(x, s)
        r = f(x_star) / f(x)
        if np.log(np.random.uniform()) < np.log(r):
            x = x_star
        samples[i] = x
    return samples
```

```python
# Insert parameters
N = 10000
s = 1
```

```python
# Generate samples
samples = metropolis_hastings(N, s)
```

```python
# Plotting the graph
plt.figure(figsize=(10, 6))
sns.histplot(samples, kde=True, bins=30, stat='density')
x_values = np.linspace(min(samples), max(samples), 1000)
plt.plot(x_values, f(x_values), 'r', label='f(x)')
plt.title('Histogram and Kernel Density Plot with f(x)')
plt.legend()
plt.show()
```

```python
# Sample mean and standard deviation
print("Sample mean:", np.mean(samples))
print("Sample standard deviation:", np.std(samples))
```
