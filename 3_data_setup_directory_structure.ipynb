```python
def analyze_data(data, year):
    # Filter out rows with NA in relevant columns
    data = data.dropna(subset=['DepDelay', 'ArrDelay', 'DepTime', 'DayOfWeek'])

    # Calculate total delay
    data['total_delay'] = data['DepDelay'] + data['ArrDelay']

    # Aggregate data by day of the week
    summary_day = data.groupby('DayOfWeek')['total_delay'].mean().reset_index(name='average_delay')

    # Find the best day
    best_day = summary_day.sort_values(by='average_delay').iloc[0]['DayOfWeek']

    # Aggregate data by hour of the day
    data['hour_of_day'] = data['DepTime'] // 100
    summary_hour = data.groupby('hour_of_day')['total_delay'].mean().reset_index(name='average_delay')

    # Find the best hour of the day (across the entire week)
    best_hour_of_day = summary_hour.sort_values(by='average_delay').iloc[0]['hour_of_day']

    return {'Year': year, 'Best_Day': best_day, 'Best_Hour_of_Day': best_hour_of_day}
```

```python
data_1990 = pd.read_csv("1990.csv")
data_1991 = pd.read_csv("1991.csv")
data_1992 = pd.read_csv("1992.csv")
data_1993 = pd.read_csv("1993.csv")
data_1994 = pd.read_csv("1994.csv")
data_1995 = pd.read_csv("1995.csv")
data_1996 = pd.read_csv("1996.csv")
data_1997 = pd.read_csv("1997.csv")
data_1998 = pd.read_csv("1998.csv")
data_1999 = pd.read_csv("1999.csv")
```

```python
results = pd.DataFrame([
    analyze_data(data_1990, 1990),
    analyze_data(data_1991, 1991),
    analyze_data(data_1992, 1992),
    analyze_data(data_1993, 1993),
    analyze_data(data_1994, 1994),
    analyze_data(data_1995, 1995),
    analyze_data(data_1996, 1996),
    analyze_data(data_1997, 1997),
    analyze_data(data_1998, 1998),
    analyze_data(data_1999, 1999)
])
```
